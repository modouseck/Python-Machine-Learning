```python
import pandas as pd

# Remplacez "votre_fichier.csv" par le nom réel de votre fichier
df = pd.read_csv("/content/Tunisair_flights_dataset.csv")

# Affiche les 5 premières lignes
print(df.head())

# Affiche un résumé des colonnes, types de données, valeurs nulles
print(df.info())

# Nombre de valeurs nulles par colonne
print(df.isnull().sum())

# Pourcentage de valeurs manquantes
print(df.isnull().mean() * 100)

# Statistiques descriptives des colonnes numériques
print(df.describe())

from ydata_profiling import ProfileReport

# Génération du rapport
profile = ProfileReport(df, title="Rapport d'exploration de données", explorative=True)

# Sauvegarde dans un fichier HTML
profile.to_file("rapport.html")





```
